```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
data_complete = pd.merge(mouse_metadata, study_results, how = "left", on = ["Mouse ID", "Mouse ID"])

# Display the data table for preview
data_complete
```

```python
# Checking the number of mice.
data_complete["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mice_duplicate = data_complete.loc[data_complete.duplicated(subset = ["Mouse ID", "Timepoint"]), "Mouse ID"].unique()

mice_duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mice_duplicate_id = data_complete.loc[data_complete["Mouse ID"] == "g989"]
mice_duplicate_id
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
data_clean = data_complete.drop(mice_duplicate_id.index)
data_clean
```

```python
# Checking the number of mice in the clean DataFrame.
data_clean["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
data_mean = data_clean["Tumor Volume (mm3)"].groupby(data_clean["Drug Regimen"]).mean()
data_median = data_clean["Tumor Volume (mm3)"].groupby(data_clean["Drug Regimen"]).median()
data_var = data_clean["Tumor Volume (mm3)"].groupby(data_clean["Drug Regimen"]).var()
data_std = data_clean["Tumor Volume (mm3)"].groupby(data_clean["Drug Regimen"]).std()
data_sem = data_clean["Tumor Volume (mm3)"].groupby(data_clean["Drug Regimen"]).sem()

# Assemble the resulting series into a single summary DataFrame.
data_summary = pd.DataFrame({
    "Mean Tumor Volume" : data_mean,
    "Median Tumor Volume" : data_median,
    "Tumor Volume Variance" : data_var,
    "Tumor Volume Std. Dev." : data_std,
    "Tumor Volume Std. Err." : data_sem
})

# Display
data_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
data_grouped = data_clean.groupby("Mouse ID")["Timepoint"].max()
data_grouped = data_grouped.reset_index().rename(columns = {"Timepoint": "timepoint_max"})

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
data_clean_grouped = data_grouped.merge(data_clean, how = "left", on = ["Mouse ID"])
data_clean_grouped
```

```python
# Put treatments into a list for for loop (and later for plot labels)
mice_treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in mice_treatments :
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    data_drug = data_clean_grouped.loc[data_clean_grouped["Drug Regimen"] == treatment]
    drug_max = data_drug.loc[data_drug["Timepoint"] == data_drug["timepoint_max"]]
    drug_vol = drug_max["Tumor Volume (mm3)"]
    
    # add subset 
    tumor_vol.append(drug_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = drug_vol.quantile([.25, .5, .75])
    low_quart = quartiles[0.25]
    high_quart = quartiles[0.75]
    range = high_quart - low_quart
    low_bound = low_quart - (1.5 * range)
    high_bound = high_quart + (1.5 * range)

    outliers = drug_vol.loc[(drug_vol < low_bound) | (drug_vol > high_bound)]

    print(f"{treatment}'s potential outliers:{outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
red_outlier = dict(markerfacecolor = "red", markersize = 10)

plt.boxplot(tumor_vol, labels =  mice_treatments, flierprops = red_outlier)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
data_capomulin = data_clean.loc[data_clean["Drug Regimen"] == "Capomulin"]
data_capomulin_avg = data_capomulin.groupby(["Mouse ID"]).mean("Tumor Size")
data_capomulin_avg.plot.scatter(x = "Weight (g)", y = "Tumor Volume (mm3)", ylabel = "Average Tumor Volume (mm3)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
mice_corr = st.pearsonr( data_capomulin_avg["Weight (g)"], data_capomulin_avg["Tumor Volume (mm3)"])

print(f"The correlation between mouse weight and the average tumor volume is {round(mice_corr[0], 2)}")

(mice_slope, mice_intercept, mice_r, mice_p, mice_std) = st.linregress(data_capomulin_avg["Weight (g)"], data_capomulin_avg["Tumor Volume (mm3)"])

mice_linregress = data_capomulin_avg["Weight (g)"] * mice_slope + mice_intercept

plt.scatter(x = data_capomulin_avg["Weight (g)"], y = data_capomulin_avg["Tumor Volume (mm3)"])
plt.plot(data_capomulin_avg["Weight (g)"], mice_linregress, "r-")
```
